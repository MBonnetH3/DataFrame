```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
ETH = pd.read_csv('Data/ETH-USD.csv',sep=','); ETH.head(3)
BNB = pd.read_csv('Data/BNB-USD.csv',sep=','); BNB.head(3)
```

```python
#data=ETH.loc[(ETH["Date"] < "2016-01-01")]
data=ETH
data["Date"] = pd.to_datetime(data["Date"])

plt.figure(figsize=(16,8))
plt.plot(data.Date,data.Low, label = 'low price')
plt.plot (data.Date,data.High, label = 'high price')
plt.xlabel('Dates')
plt.ylabel('Price ($)')
plt.title('volution de l Etherum')
plt.legend();
```

```python
sns.relplot(x="Volume", y="Low", data = ETH, height=6);
plt.title('Le prix le plus bas en fonction du volume de l etherum');
```

```python
data=ETH.loc[((ETH["Date"] > "2021-01-01") & (ETH["Date"] < "2021-01-10"))]
#data=ETH
data["Date"] = pd.to_datetime(data["Date"])

plt.figure(figsize=(12,8))
plt.scatter(data.Date, data.Open,c='Green',label = 'Open')
plt.scatter(data.Date, data.Close,c='Red',label = 'Close')
plt.xlabel('Nombre de lits')
plt.ylabel('Date')
plt.legend()
plt.title('L ouverture et la fermeture de l Ethereum au mois de janvier 2021');
```

```python
data=ETH.loc[((ETH["Date"] > "2020-08-31") & (ETH["Date"] < "2020-10-10"))]
#data=ETH
data["Date"] = pd.to_datetime(data["Date"])

plt.figure(figsize=(12,8))
plt.scatter(data.Date, data.Open,c='Green',label = 'Open')
plt.scatter(data.Date, data.Close,c='Red',label = 'Close')
plt.xlabel('Nombre de lits')
plt.ylabel('Date')
plt.legend()
plt.title('L ouverture et la fermeture de l Ethereum au mois de janvier 2020 ');
```

```python

plt.figure(figsize=(12,7))
ETH['Cours'].value_counts().plot(kind='barh')
plt.title("Nombre de cours positif ou negatif")
```
